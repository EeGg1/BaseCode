```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("./data/crypto_binance_spot_1h/BTCUSDT.csv")
data.shape
```

```python
length = 600
linewidth = 0.7

plt.figure(figsize=(12.2, 4.5))
plt.plot(data['Close'][15:length], label='Close', linewidth=linewidth)
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(12.2, 4.5))
plt.plot(data['StochRSI'][15:length], label='StochRSI', linewidth=linewidth)
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(12.2, 4.5))
plt.plot(data['MACD'][15:length], label='MACD', linewidth=linewidth)
plt.plot(data['MACD_signal'][15:length], label='MACD_signal', linewidth=linewidth)
plt.plot(data['MACD_hist'][15:length], label='MACD_hist', linewidth=linewidth)
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np

test_worst_enc_window = np.load('data_all[test_worst_enc_window].npy')
test_worst_ground_truth = np.load('data_all[test_worst_ground_truth].npy')
test_worst_pred = np.load('data_all[test_worst_pred].npy')
```

```python
print(test_worst_enc_window.shape)
print(test_worst_ground_truth.shape)
print(test_worst_pred.shape)
```

```python
import matplotlib.pyplot as plt
import os

for i in range(test_worst_enc_window.shape[1]):
    plt.clf()  
    plt.plot(test_worst_enc_window[:,i], label='input', color='green')
    plt.plot(range(len(test_worst_enc_window), len(test_worst_enc_window)+len(test_worst_ground_truth)),test_worst_ground_truth[:,i], label='ground truth', color='red')
    plt.plot(range(len(test_worst_enc_window), len(test_worst_enc_window)+len(test_worst_pred)),test_worst_pred[:,i], label='pred', color='blue')
    plt.legend(loc='upper left')
    
    if not os.path.exists('./fig'):
        os.makedirs('./fig')
    save_path = os.path.join('./fig', f'test_worst_{i+1}.png')
    plt.savefig(save_path)
    
```
